```python
# Streamlit version of the CyberSec Toolkit: CTF & CEH Edition
import streamlit as st
import subprocess
import os
import datetime
from pathlib import Path

st.set_page_config(page_title="CyberSec Toolkit", layout="wide")
st.title("🛡️ CyberSec Toolkit: CTF & CEH Edition")

def create_output_folder():
    timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
    folder = Path(f"session_logs/session_{timestamp}")
    folder.mkdir(parents=True, exist_ok=True)
    return folder

def run_command(cmd):
    folder = create_output_folder()
    try:
        result = subprocess.check_output(cmd, shell=True, stderr=subprocess.STDOUT, text=True)
    except subprocess.CalledProcessError as e:
        result = e.output
    log_path = folder / "output.txt"
    log_path.write_text(result)
    return result, log_path

section = st.sidebar.radio("Navigate", ["CTF Toolkit", "CEH Modules", "Notes"])

if section == "CTF Toolkit":
    st.header("🎯 CTF Modules")
    ctf_tabs = st.tabs(["Crypto", "Stego", "Reversing", "Web"])

    with ctf_tabs[0]:
        st.subheader("🔐 Crypto Tools")
        text_input = st.text_input("Enter text to decode")
        col1, col2, col3 = st.columns(3)
        if col1.button("Base64 Decode"):
            cmd = f"echo {text_input} | base64 -d"
            output, path = run_command(cmd)
            st.code(output)
        if col2.button("ROT13 Decode"):
            cmd = f"echo {text_input} | tr 'A-Za-z' 'N-ZA-Mn-za-m'"
            output, path = run_command(cmd)
            st.code(output)
        if col3.button("XOR Brute-force File"):
            xor_file = st.file_uploader("Upload Encrypted File")
            if xor_file:
                data = xor_file.read()
                folder = create_output_folder()
                result = []
                for key in range(256):
                    decoded = bytes([b ^ key for b in data])
                    try:
                        text = decoded.decode("utf-8")
                        if all(32 <= ord(c) < 127 or c in '\n\r\t' for c in text[:100]):
                            result.append(f"[KEY {key}] {text[:200]}\n")
                    except:
                        continue
                out = "\n".join(result) if result else "No readable content."
                (folder / "xor_output.txt").write_text(out)
                st.code(out)

    with ctf_tabs[1]:
        st.subheader("🖼️ Steganography Tools")
        password = st.text_input("Steghide Password", type="password")
        if st.button("Run Steghide"):
            uploaded_file = st.file_uploader("Upload stego file")
            if uploaded_file:
                with open("temp.jpg", "wb") as f:
                    f.write(uploaded_file.read())
                cmd = f"steghide extract -sf temp.jpg -p {password} -xf extracted_output.txt"
                output, path = run_command(cmd)
                st.code(output)
        if st.button("Run ExifTool"):
            image = st.file_uploader("Upload image")
            if image:
                with open("meta.jpg", "wb") as f:
                    f.write(image.read())
                cmd = f"exiftool meta.jpg"
                output, path = run_command(cmd)
                st.code(output)
        if st.button("Run Zsteg"):
            image = st.file_uploader("Upload PNG/BMP")
            if image:
                with open("zimg.png", "wb") as f:
                    f.write(image.read())
                cmd = f"zsteg zimg.png"
                output, path = run_command(cmd)
                st.code(output)

    with ctf_tabs[2]:
        st.subheader("⚙️ Reversing Tools")
        binary = st.file_uploader("Upload binary file")
        col1, col2, col3 = st.columns(3)
        if binary:
            path = Path("reversetemp.bin")
            path.write_bytes(binary.read())
            if col1.button("Strings"):
                cmd = f"strings {path}"
                output, log = run_command(cmd)
                st.code(output)
            if col2.button("Binwalk"):
                cmd = f"binwalk {path}"
                output, log = run_command(cmd)
                st.code(output)
            if col3.button("Radare2 Main"):
                cmd = f"r2 -A {path} -c 'aa; pdf @ main'"
                output, log = run_command(cmd)
                st.code(output)

    with ctf_tabs[3]:
        st.subheader("🌐 Web Exploitation")
        target = st.text_input("Target URL/IP")
        col1, col2, col3 = st.columns(3)
        if col1.button("SQLMap"):
            cmd = f"python3 sqlmap/sqlmap.py -u \\\"{target}\\\" --batch"
            output, path = run_command(cmd)
            st.code(output)

        if col2.button("Nmap"):
            cmd = f"nmap -sV {target}"
            output, path = run_command(cmd)
            st.code(output)
        if col3.button("Dirsearch"):
            cmd = f"dirsearch -u \"{target}\""
            output, path = run_command(cmd)
            st.code(output)

elif section == "CEH Modules":
    st.header("🧠 CEH Modules")
    ceh_tabs = st.tabs(["Reconnaissance", "Enumeration", "Password Attacks", "Web App Testing"])

    with ceh_tabs[0]:
        st.subheader("🔍 Reconnaissance Tools")
        domain = st.text_input("Enter Domain/IP")
        if st.button("WHOIS Lookup"):
            output, path = run_command(f"whois {domain}")
            st.code(output)
        if st.button("NSLookup"):
            output, path = run_command(f"nslookup {domain}")
            st.code(output)

    with ceh_tabs[1]:
        st.subheader("🔎 Enumeration Tools")
        target_enum = st.text_input("Target IP for Enum4Linux")
        if st.button("Run Enum4Linux"):
            cmd = f"enum4linux -a {target_enum}"
            output, path = run_command(cmd)
            st.code(output)

    with ceh_tabs[2]:
        st.subheader("🔐 Password Attack Tools")
        ssh_user = st.text_input("SSH Username")
        ssh_host = st.text_input("Target IP")
        wordlist = st.text_input("Path to Wordlist (e.g., /usr/share/wordlists/rockyou.txt)")
        if st.button("Run Hydra (SSH)"):
            cmd = f"hydra -l {ssh_user} -P {wordlist} ssh://{ssh_host}"
            output, path = run_command(cmd)
            st.code(output)

    with ceh_tabs[3]:
        st.subheader("🌐 Web Application Testing")
        url = st.text_input("Enter Target URL")
        if st.button("Nikto Scan"):
            cmd = f"nikto -h {url}"
            output, path = run_command(cmd)
            st.code(output)

elif section == "Notes":
    st.header("📝 Challenge Notes")
    notes = st.text_area("Enter notes for this session")
    if st.button("Save Notes"):
        folder = create_output_folder()
        path = folder / "notes.txt"
        path.write_text(notes)
        st.success(f"Notes saved to {path}")
```
